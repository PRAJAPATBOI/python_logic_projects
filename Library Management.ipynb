```python
books = [ 'PYTHON', 'C LANGUAGE', 'MATERIAL FLUID', 'DSA', 'CSO', 'OS'] #List to store data of books
students = ['RAMESH', 'DINESH', 'ANURAG', 'MADHAV', 'CHETAN'] #List to store data about students
booksdata = [{'C++' : 'RAJ'}]

while True :
    print("1. Add a book")
    print("2. Borrow a book")
    print("3. Submit a book")
    print("4. Add a student")
    print("5. Print the books")
    print("6. Print the students")
    print("7. Print the details of borrowed books")
    print("8. Quit")
    a = int(input("Enter your choice : "))
    
    if a == 1 :
        t1 = input("Enter the name of book : ").upper()
        books.append(t1)
    elif a == 2 :
        t1 = input("Enter the name of book you want to borrow : ").upper()
        if books.count(t1) >= 1 :
            st = input("Enter name of student : ").upper()
            if students.count(st) >= 1 :
                books.remove(t1)
                t3 = {t1:st} #Book's name will be stored as key whereas value will be student's name
                booksdata.append(t3)
                print(f"You have successfully borrowed {t1} book.")
            else :
                print("Invalid Input / Register First / Retry")
        else :
            for i in booksdata :
                t5 = "dict_keys([\'" + t1 + "\'])"
                t6 = str(i.keys())
                if t6 == t5 :
                    print(f"This book is already borrowed by : {i}")
                    #t2 = booksdata.index(t1)
                    #print(f"This book is already borrowed by : {t2[t1]}")
                elif i == booksdata[len(booksdata)-1] :
                    print("Your requested book is not available")
    elif a == 3 :
        sb = input("Enter name of book you want to submit : ").upper()
        for i in booksdata :
            tt1 = str(i.keys()).strip("dict_keys([''])")
            if sb == tt1 :
                booksdata.remove(i)
                books.append(sb)
                print("You have successfully submitted the book")
            else:
                print(tt1)
                print(sb)
    elif a == 4 :
        t1 = input("Enter the name of student : ").upper()
        students.append(t1)
    elif a == 5 :
        print(books)
    elif a == 6 :
        print(students)
    elif a == 7 :
        print(booksdata)
    elif a == 8 :
        print("Thanks Visit Again")
        break
    else :
        print()
        print("Invalid Input")
        print("Try Again")
        print()
```
