```python
import math
class area: #Class for 1 Choice
    def square(self,l):
        return l**2
    def rectangle(self,l,b):
        return l*b
    def circle(self, r):
        return 3.14159*r*r
    def triangle(self,a,b,c):
        s = (a + b + c)/2
        return math.sqrt(s*(s-a)*(s-b)*(s-c))
class perimeter: #Class for 1 Choice
    def square(self,l):
        return 4*l
    def rectangle(self,l,b):
        return 2*(l+b)
    def circle(self, r):
        return 2*3.14159*r
    def triangle(self,a,b,c):
        return a+b+c
    

class surface_area: #Class for 2 choice
    def cube(self,s):
        return 6*s**2
    def r_circular_cylinder(self,r,h):
        return 2*3.14159*r*(r + h)
    def solid_sphere(self,r):
        return 4*3.14159*r*r
    def hemisphere(self,r):
        return 3*3.14159*r*r
class volume: #Class for 2 choice
    def cube(self,s):
        return s*s*s
    def r_circular_cylinder(self,r,h):
        return 3.14159*r*r*h
    def solid_sphere(self,r):
        return 4/3*3.14159*r*r*r
    def hemisphere(self,r):
        return 2/3*3.14159*r*r*r

class temperatureconverter: #Class for 3 choice
    def celsius_to_kelvin(self,c):
        return c+273.15
    def kelvin_to_celsius(self,k):
        return k-273.15
    def fahrenheit_to_celsius(self,f): 
        return (f-32)*(5/9)
    def celsius_to_fahrenheit(self,c):
        return c*(9/5)+32


ar = area() #object creation for area class
pr = perimeter() #object creation for perimeter class
sr = surface_area() #object creation for a surface area class
vl = volume() #object creation for a volume class
tmp = temperatureconverter() #object creation for temperatureconverter class


print('Application designed to calculate different Mathematical calculations')
print('1.Find Area and Perimeter of different Shapes\n2.Find Suface Area and Volume of Solid Shapes\n3.Temprature Conversions\n4.Exit Application')
print()

try:
    while True:
        choice = int(input('Enter your choice among 3 : '))
        if choice == 1:
            print('Section to find Area and Perimeter of different Shapes')
            print('1.Square\n2.Rectangle\n3.Circle\n4.Triangle\n5.Exit')
            while True:
                choice = int(input('Enter your choice : '))
                if choice == 1:
                    length = float(input('Enter the side of Square : '))
                    print(f"Area of Square with side {length} = {ar.square(length):.2f}")
                    print(f"Perimeter of Square with side {length} = {pr.square(length):.2f}")
                elif choice == 2:
                    length = float(input('Enter the length of Rectangle : '))
                    breadth = float(input('Enter the breadth of Rectangle : '))
                    print(f"Area of Rectangle with length {length} and {breadth} = {ar.rectangle(length,breadth):.2f}")
                    print(f"Perimeter of Rectangle with length {length} and {breadth} = {pr.rectangle(length,breadth):.2f}")
                elif choice == 3:
                    radius = float(input('Enter the radius of a circle : '))
                    print(f"Area of a circle with radius {radius} = {ar.circle(radius):.2f}")
                    print(f"Circumference of a circle with radius {radius} = {pr.circle(radius):.2f}")
                elif choice == 4:
                    side1 = float(input('Enter the 1st side of a triangle : '))
                    side2 = float(input('Enter the 2nd side of a triangle : '))
                    side3 = float(input('Enter the 3rd side of a triangle : '))
                    print(f"Area of a triange with sides {side1}, {side2} and {side3} = {ar.triangle(side1,side2,side3):.2f}")
                    print(f"Perimeter of a triange with sides {side1}, {side2} and {side3} = {pr.triangle(side1,side2,side3):.2f}")
                elif choice == 5:
                    break
                else:
                    print('Invalid Choice\nPlease Retry')
        elif choice == 2:
            print('Section to find Surface Area and Volume of Solid Shapes')
            print('1.Cube\n2.Right Circular Cylinder\n3.Solid Sphere\n4.Hemisphere\n5.Exit')
            while True:
                choice = int(input('Enter your choice : '))
                if choice == 1:
                    side = float(input('Enter the side of a cube : '))
                    print(f'Total Surface Area of a cube with side {side} = {sr.cube(side):.2f}')
                    print(f'Volume of a cube with side {side} = {vl.cube(side):.2f}')
                elif choice == 2:
                    radius = float(input('Enter radius of the Right Circular Cylinder : '))
                    height = float(input('Enter height of the Right Circular Cylinder : '))
                    print(f'Total Surface Area of a right circular cylinder with radius {radius} and height {height} = {sr.r_circular_cylinder(radius,height):.2f}')
                    print(f'Volume of a right circular cylinder with radius {radius} and height {height} = {vl.r_circular_cylinder(radius,height):.2f}')
                elif choice == 3:
                    radius = float(input('Enter radius of the Solid Sphere : '))
                    print(f'Total Surface Area of a solid sphere with radius {radius} = {sr.solid_sphere(radius):.2f}')
                    print(f'Area of a solid sphere with radius {radius} = {vl.solid_sphere(radius):.2f}')
                elif choice == 4:
                    radius = float(input('Enter radius of the Hemisphere : '))
                    print(f'Total Surface Area of a Hemisphere with radius {radius} = {sr.hemisphere(radius):.2f}')
                    print(f'Volume of a Hemisphere with radius {radius} = {vl.hemisphere(radius):.2f}')
                elif choice == 5:
                    break
                else:
                    print('Invalid Choice\nPlease Retry')
        elif choice == 3:
            print('Section to Convert the Temprature')
            print(f'1.Celsius to Kelvin\n2.Kelvin to Celcius\n3.Fahrenheit to Celcius\n4.Celsius to Fahrenheit\n5.Exit')
            while True:
                choice = int(input('Enter your choice : '))
                if choice == 1:
                    celsius = float(input('Enter temprature in celsius : '))
                    print(f'Temprature in Kelvin = {tmp.celsius_to_kelvin(celsius):.2f}')
                elif choice == 2:
                    kelvin = float(input('Enter temprature in kelvin : '))
                    print(f'Temprature in Celsius = {tmp.kelvin_to_celsius(kelvin):.2f}')
                elif choice == 3:
                    fahrenheit = float(input('Enter temprature in fahrenheit : '))
                    print(f'Temprature in Celsius = {tmp.fahrenheit_to_celsius(fahrenheit):.2f}')
                elif choice == 4:
                    celsius = float(input('Enter temprature in celsius : '))
                    print(f'Temprature in Fahrenheit = {tmp.celsius_to_fahrenheit(celsius):.2f}')
                elif choice == 5:
                    break
                else:
                    print('Invalid Choice\nPlease Retry')
        elif choice == 4:
            break
        else:
            print('Invalid Choice\nPlease Retry')
except ValueError:
    print('Please enter an non-string number')
except Exception as e:
    print('Please enter correct values')
    print(f'Error = {e}')
```
